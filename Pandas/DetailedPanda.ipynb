```python
# delete multiple columns
df.drop(['Married','Weight'],axis=1,inplace=True)
df.head()
```

```python
# # to check how many null values are there in column
df.isnull().sum()
```

```python
# Now we play with rows
```

```python
# Rows selection using loc[]

num = df.loc[0,'Age']
print(num)
```

```python
# Select Multiple rows and columns

num = df.loc[0:4,['Name','Team','Age']]
num
```

```python
# all rows and multiple columns

num = df.loc[:,['Name','Team','Age']]
num
```

```python
# Boolean Indexing -> Filters rows based on condition that evaluates to a boolean(True or False)
filtered_data = df[df['Age']%2==0]
filtered_data.head()
```

```python
# Multiple Conditions

filtered_data = df[(df['Age']%2==0) & (df['Age']>22)]
filtered_data.head()
```
