```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('titanic.txt', index_col='PassengerId')
data
```

```python
data['Survived'] = np.where(data['Survived'] == 1,'Yes','No')
data
```

```python
data['Age'] = data['Age'].mode()[0].astype(int)
data
```

```python
data.head(10)
```

```python
data['Embarked'] = data['Embarked'].apply(lambda x:'Southampton, England' if x == 'S'
                                             else 'Cherbourg, France' if x == 'C'
                                             else 'Queenstown, Ireland' )

data.head(50)
```

```python
data.describe()
```

```python
data["Married"] = "Unknown"
data
```

```python
def check_married(group):
    if 'male' in group['Sex'] and 'female' in group['Sex']:
        group[Married] = "Yes"
    else:
        group[Married] = "No"      
    return group    

data = data.groupby('Cabin').apply(check_married)

data    
```
