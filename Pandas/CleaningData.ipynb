```python
data = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva', 'Alice', 'Bob'],
    'Age': [22, -30, 22, np.nan, 28, 22, np.nan],  # Negative and missing age
    'Salary': [50000, 55000, np.nan, 60000, np.nan, 50000, 55000],  # Missing and duplicate salary
    'Date': ['1996-05-14', 'not_a_date', '1999-02-22', '1990-11-10', '1987-07-25', '1996-05-14', '1992-03-17'],  # Wrong format and duplicate date
    'City': ['New York', 'San Francisco', np.nan, 'Los Angeles', 'Chicago', 'New York', np.nan]  # Missing city and duplicate
})
print(data.to_string())
print(type(data))
```

```python
# num = data['Age'].mean()
num = data['Age'].median()
# num = data['Age'].mode()[0]
data.fillna(num, inplace = True)
print(data.to_string())
```

```python
data.dropna(subset=['Date'],inplace = True)
print(data)
```

```python
for x in data.index:
    if data.loc[x,"Salary"] < 50000:
        data.loc[x,"Salary"] = 50000

print(data)
```

```python
# remove duplicates
data.drop_duplicates(inplace = True)
print(data)
```

```python
# to fill value with their avg

# for numeric values
for column in data.select_dtypes(include=['float64','int64']).columns:
    data[column].fillna(data[column].mean(), inplace = True)

# for City
data['City'].fillna(data['City'].mode()[0], inplace = True)

# Convert 'Date' column to datetime, invalid entries will become NaT
data['Date'] = pd.to_datetime(data['Date'], errors='coerce')

# Replace NaT (invalid dates) in 'Date' column with the mode (most frequent valid date)
data['Date'].fillna(data['Date'].mode()[0], inplace=True)

data
```

```python
# adding a new column 
data['Married'] = 'False'
data
```

```python
data['Age'] = data['Age'].abs()
data
```
